> **Principal Component Analysis**

```python
mydata<-read.csv('Data Survei2.csv')
```

```python
mydata
```

```python
attach(mydata)
```

```python
# Define Variabels
x <-cbind(Q1,Q2,Q3,Q4,Q5,Q6,Q7,Q8)
```

```python
# Descriptive statistics
summary(x)
cor(x)
```

```python
# principal component analysis
pca <-princomp(x,score=TRUE, cor=TRUE)
```

```python
# loading of principal components
loadings(pca)
```

> **Menampilkan Grafik Hasil Uji Data Dengan PCA**

```python
# scree plot of eigenvalues 
plot(pca)
screeplot(pca, type="line",
main="scree plot")
```

*Dari grafik diatas, disimpulkan bahwa Comp. 1 dan 2/pertanyaan 1 dan 2 memiliki varian lebih tinggi dari yang lainnya, sehingga kedua komponen ini yang menjadi layak untuk dianalisa.*

```python
# Biplot of score variables
biplot(pca)
```
